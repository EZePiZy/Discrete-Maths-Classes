```python
def unique_items(X):
    U = []
    for x in X:
        has_it = False
        for u in U:
            if u==x:
                has_it = True
                break
        if not has_it:
            U.append(x)
    return U
```

```python

# Python program to check if two
# to get unique values from list
# using set
  
# function to get unique values
  
  
def unique(list1):
  
  
    output = list(set())
    for x in list1: # n
        if x not in output: # O(1)
            output.append(x) # O(1)
    return output # O(1)

```

```python
def unique_np(list1):
    x = np.array(list1)
    return str(np.unique(x))
```

```python
 
def is_unique2 (alist : [int]) -> list:
    U = set() # O(1) - Initialising variable
    alist = set(alist) # O(N) - Changing data structure 
    copy = sorted(alist)  # O(N Log N) - for fast Python sorting
    for i in range(len(alist)-1): # O(N) - really N-1, but that is O(N); len and - are both O(1)
        if copy[i] == copy[i+1]: # O(1): +, 2 [i], and  == on ints: all O(1)
            U.add(copy[i+1]) # O(1)
    U = alist - U # O(N) - always executed: O(N) because only have to iterate through all elements of alist once. For each elements Set U is tested for membership in O(1)
    return list(U) # O(1) - always executed in worst case

# Total complexity = O(1) + O(N) + O(N Log N) + O(N)*O(1) + O(N) + O(1) = O(N + N Log N + O(N*1) + N + 1) = O(N + N Log N + N + N + 1) = O(N Log N + 3N + 1) = O(N Log N)
```

```python
def set_test(alist : [int]) -> list:
    U = set()
    for i in range(len(alist)):
        U.add(alist(i))
    return list(U)

lst = [random.choice(range(100)) for i in range(1000)]

print(set_test(lst))
```

```python
all_list = []
for x in range(5000):
    lst = [random.choice(range(100)) for i in range(x)]
    all_list.append(lst)

#print(all_list)

#lst = [random.choice(range(100)) for i in range(300)]
#print(len(lst))
```

```python
def timeit(n, repeat):
    start = time.time()
    for i in range(repeat):
        is_unique2(n)
    end = time.time()
    return (end - start) / repeat

def timeit_2(n, repeat):
    start = time.time()
    for i in range(repeat):
        unique_items(n)
    end = time.time()
    return (end - start) / repeat

t_new_func = [timeit(n, 1) for n in all_list]
t_old_func = [timeit_2(n, 1) for n in all_list]


fig, ax = plt.subplots()

plt.plot(range(len(all_list)), t_new_func, label="New function", color='green')
plt.plot(range(len(all_list)), t_old_func, label="Old function" , color='red')
plt.xlabel('Number of elements in randomly genereated list')
plt.ylabel('Time (s)')
plt.legend()





```
