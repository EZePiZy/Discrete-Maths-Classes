```python
#Creating test array:
test = np.random.choice(a=[False, True], size=300)
```

```python
#Naive algorithm: 
def bool_sort(lst):
    lst_sorted = [None] * (len(lst) + 1) 
    count_T = 1
    count_F = 1

    for i in range(len(lst)):
        if (lst[i] == False):
            lst_sorted[count_F] = False
            count_F += 1
        else:
            lst_sorted[len(lst) - count_T + 1] = True
            count_T += 1 

    lst_sorted.pop(0)   
    return ''''The sorted list is: ' + str(lst_sorted),''' 'The number of occurences of "False" is: ' + str(count_F - 1), 'The number of occurences of "True" is: ' + str(count_T - 1)
```

```python
#Testing Naive algorithm: 
print(bool_sort(test))
```

```python
#Divide and Conquer algorithm: 

def dac_bool(lst, low = 0, high=None ):
    if high is None:
        high = len(lst) - 1 # Setting top boundary - O(1)
    
    if low == high: # O(1)
        return lst
    if lst[low] == False: 
        return dac_bool(lst, low + 1, high)
    else: 
        lst[low] = lst[high]
        lst[high] = True
        return dac_bool(lst, low, high - 1)
  

   
```

```python
def timeit_naive(n, repeat):
    start = time.time()
    for i in range(repeat):
        bool_sort(n)
        
    end = time.time()
    return (end - start) / repeat

T_sort = [timeit_naive(n, 100) for n in test]


plt.figure(figsize=(12, 4))
plt.subplot(121)
plt.plot(range(len(test)), T_sort)
plt.xlabel('n')
plt.ylabel('Time (us)')
```
