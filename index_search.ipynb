```python
def index_search(lst, x, low=0, high=None):
    if high is None:
        high = len(lst) - 1
    
    if high >= low:
        mid = low + (high - low) // 2
        
        if lst[mid] == x:
            return mid 
        
        elif lst[mid] > x:
            return index_search(lst, x, low, mid - 1)
        
        else: 
            return index_search(lst, x, mid + 1, high)
    
    else: 
        return -1


# n * (1/2)k = 1\newline
# n * 1/2k = 1\newline
# n = 2k\newline
# log2(n) = k → O(log(n))

# ■ QED
```
