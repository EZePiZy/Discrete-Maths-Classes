```python
import math
from collections import defaultdict
import graphviz
infinity = float('inf')
```

```python
def visualise_weighted_graph(G, W):
    dot = graphviz.Graph()
    dot.attr(rankdir='LR')
#     print('keys: ')
#     print(G.keys())
#     print('items: ')
#     print(W.items())
    for name in G.keys():
        dot.node(str(name))
    for (u, v), w in W.items():
        if u < v:
            dot.edge(str(u), str(v), label=str(w))
    return dot

```

```python
def design_graph():
    G = defaultdict(list)
    W = {}
    nodes = [x for x in range(2, 51)]
    for i in nodes: 
        for j in nodes: 
            if i != j:
                if math.gcd(i, j) > 1:
                    G[i].append(j)
                    G[j].append(i)
                    W[i, j] = W[j, i] = 1/math.gcd(i,j)
                
    return G, W

# G, W = design_graph()
print(design_graph())
# visualise_weighted_graph(G,W)
```

```python
start = 9
predecessor = {start: None}
q = {start: 0}
shortest = {}
```

```python
def djikstra(G, W):
        final = 0
        while q:
            d, vertex = min((d, vertex) for vertex, d in q.items())
            print(d, vertex)
#             del q[vertex]

            shortest[vertex] = d
            final += d

            if vertex == 50:
                return final

            for next_vertex in G[vertex]:
                # print(next_vertex)
                if next_vertex in shortest:
                    continue
                c = q.get(next_vertex, infinity)
                e = d+W[vertex, next_vertex]
                if e < c:
                    q[next_vertex] = e
                    predecessor[next_vertex] = vertex
                    
                    
        return final
```

```python
G, W = design_graph()
print(djikstra(G, W))
# visualise_weighted_graph(G,W)
```
